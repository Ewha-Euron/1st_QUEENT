```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
def date_range(start, end):
    start = datetime.strptime(start, "%Y/%m/%d")
    end = datetime.strptime(end, "%Y/%m/%d")
    dates = [(start + timedelta(days=i)).strftime("%Y/%m/%d") for i in range((end-start).days+1)]
    return dates

dates = date_range("2010/07/01", "2021/12/27")

def slash_to_comma(x):
    x = x.replace('-','/')
    return x

def dot_to_slash(x):
    x = x.replace('.','/')
    return x

def slash_reorder(x):
    x = x[6:]+'/'+x[:2]+'/'+x[3:5]
    return x

def slash_reorder_yr(x):
    x = '20'+x[6:]+'/'+x[0:2]+'/'+x[3:5]
    return x

def gold_date(x):
    li = x.split('.')
    x = li[0]+'/'+li[1].zfill(2)+'/'+li[2].zfill(2)
    return x
# gold_date('2010.1.22')
```
